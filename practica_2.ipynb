plotting sin linea

para trazar solo los marcadores, puedes usar notacion de cadena de acceso directo 'o', que significa "anillos"

ejemplo:dibuja dos puntos en el diagrama, uno en la posicion(1,3) y otro en la posicion(8,10)

```python
import numpy as np
import matplotlib.pyplot as plt

puntosX = np.array([1, 8])
puntosY = np.array([3, 10])

plt.plot(puntosX, puntosY, 'o')
plt.show()
```

diagramas de dispersion (scatter)

se emplean para mostrar la relacion entre dos variables numericas,detectando corrrelacciones o agrupamientos

ejemplo:


```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
peso = [50,60,65,80,85,95]

plt.scatter(altura, peso, color = 'red', marker = 'x')
plt.title('relacion entre la Altura yPeso')
plt.xlabel('Altura (cms)')
plt.ylabel('Peso(kg)')
plt.show()
```

graficos de barras (bar yb barh)

sirve para comparar categorias o variables tanto en orientacion vertical como en horizontal

ejemplo de barras en vertical:

```python
import matplotlib.pyplot as plt

categorias = ["A", "B", "C", "D"]
valores = [4, 7, 1, 8]

plt.bar(categorias, valores, color ="orange")
plt.title('Grafico de barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```
